```python
for ducks in os.listdir(r'duck'):
    duck = cv2.imread(ducks, cv2.IMREAD_COLOR)
    #print(duck)
    #duck = Image.open(ducks)
    width = duck.shape[0]
    height = duck.shape[1]
    #duck = duck.convert('RGB')
    duckarray = []
    ducklabel=[]
    #print(duck)
    for x in range(width):
        for y in range(height):
            duckarray.append(duck[x,y])
            ducklabel.append(1)
            #print(duck)
            #print(duckarray)
#print (np.mean(duckarray))
#print (np.cov(duckarray))
#print(duckarray)
```

```python
for unducks in os.listdir(r'unduck'):
    unduck = cv2.imread(unducks, cv2.IMREAD_COLOR)
    #unduck = Image.open(unducks)
    width = unduck.shape[0]
    height = unduck.shape[1]
    #unduck = unduck.convert('RGB')
    unduckarray = []
    unducklabel=[]
    #print(unduck)
    for x in range(width):
        for y in range(height):
            unduckarray.append(unduck[x,y])
            unducklabel.append(0)
```

```python
GNB = GaussianNB()
GNB.fit(duckarray, ducklabel)
```

```python
findduck = cv2.imread('full_duck.jpg', cv2.IMREAD_COLOR)
width = findduck.shape[0]
height = findduck.shape[1]
for x in range(width):
        data=[]
        
        for y in range(height) :
            data.append(findduck[x,y])

        data = np.array(data)
        predict = GNB.predict(data)
            
        for y in range(height) :
            if(predict[y]==1):
                findduck[x,y]=0,0,255
```
